```python
from numpy import average
import pandas as pd

data = {
    "Student": ["John", "Mary", "David", "Sarah", "Daniel"],
    "Math": [85, 90, 78, 65, 95],
    "English": [80, 85, 75, 70, 98],
    "Science": [88, 92, 80, 60, 96]
}

df = pd.DataFrame(data)
# print out the average scores from each subjects
df["average"] = (
    df["Math"] +
    df["English"] +
    df ["Science"]
) /3

# This shows the best student
best_student = df.loc[df["average"].idxmax()]

# This shows the student that failed the most
failed_student = df[df["average"]<70]

# This arranges the student score in adescending manner
sorted_student = df.sort_values(by ="average", ascending=False)

# represent it graphically
import matplotlib.pyplot as plt
plt.bar(df["Student"], df["average"])
plt.title("Student Average Scores")
plt.xlabel("Student")
plt.ylabel("Average scores")

plt.show()

print(best_student)
print(failed_student)
print(sorted_student)
```
